# 1. Problem Statement

We are trying to predict carbon emissions in Rwanda (country in Africa) based on historical data. The data is based on 3 years of data from 2019 - 2021. We have 7 main features including:
- Sulphur Dioxide
- Carbon Monoxide
- Nitrogen Dioxide
- Formaldehyde
- UV Aerosol Index
- Ozone 
- Cloud

Each of these features have sub-features that was calculated using DOAS technique. Furthermore, there are 497 unique locations that were selected from areas in Rwanda.

```python
def remove_null_cols(data):
    df = data.copy()
    cols_drop1 = null_df_test.sort_values('Total', ascending=False).index.to_list()[:7]
    cols_drop2 = null_df_train.sort_values('Total', ascending=False).index.to_list()[:7]
    assert cols_drop1 == cols_drop2, "Features are not identical" # check if we have the same features name
    df.drop(cols_drop1, axis=1, inplace=True)
    
    return df
```

## Encoding latitude + longitude as location

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def preprocess_data(data):
    df = data.copy()
    df = remove_null_cols(df)
    if 'location' not in train_data.columns.to_list():
        train_data['location'] = train_data['latitude'].astype(str) + ' ' + df['longitude'].astype(str)
    df['location'] = df['latitude'].astype(str) + ' ' + df['longitude'].astype(str)
    enc = LabelEncoder()
    enc.fit(train_data['location'])
    df['location'] = enc.transform(df['location'])
    return df
```

### Helper Functions

```python
from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras.layers import Dense
import tensorflow.keras.optimizers as tfo
from tensorflow.keras import Sequential
```

```python
def scale_data(X_train, X_val):
    scaler = StandardScaler()
    s = scaler.fit(X_train)
    Xt = s.transform(X_train)
    Xv = s.transform(X_val)
    return Xt, Xv
```

```python
# preprocessing
def split():
    train_df1 = preprocess_data(train_data)
    features = train_df1.drop(['emission','ID_LAT_LON_YEAR_WEEK', 'latitude', 'longitude'], axis=1)
    target = train_df1.emission
    
#     train_split = train_df1[train_df1.year!=2021]
#     val_split = train_df1[train_df1.year==2021]
    X_train, X_val, y_train, y_val = train_test_split(features, target, test_size=.2, random_state=5)
#     X_train, y_train, X_val, y_val = train_split[features], train_split[target], val_split[features], val_split[target]
    
    X_train.fillna(0, inplace=True)
    X_val.fillna(0, inplace=True)
    
    X_train = tf.Variable(X_train.values, dtype=tf.float32, name='X_train')
    X_val = tf.Variable(X_val.values, dtype=tf.float32, name='X_val')
    y_train = tf.constant(y_train.values, dtype=tf.float32, name='y_train')
    y_val = tf.constant(y_val.values, dtype=tf.float32, name='y_val')
    
    X_train, X_val = scale_data(X_train, X_val)
    return X_train, X_val, y_train, y_val
```

```python
X_train, X_val, y_train, y_val = split()
```

# 3. Trying Neural Network using Keras

```python
def plot_loss(history):
    train_loss = history['loss']
    val_loss = history['val_loss']
    epochs = range(1, len(train_loss)+1)
    plt.plot(epochs, train_loss, 'bo', label='Training Loss')
    plt.plot(epochs, val_loss, 'b', label='Validation Loss')
    plt.title("Training and validation loss")
    plt.xlabel("epochs")
    plt.ylabel("Loss")
    plt.legend()
```

```python
from tensorflow.config.experimental import list_physical_devices
list_physical_devices()
```

```python
def build_NN(n_hidden_layers, n_units):
    model = Sequential()
    model.add(Dense(n_units[0], activation='relu', input_shape=(X_train.shape[1],), 
#                    kernel_regularizer=tf.keras.regularizers.L1(0.01),
#                    activity_regularizer=tf.keras.regularizers.L2(0.01)
                   ))
    if n_hidden_layers>1:
        for i in range(1,n_hidden_layers):
            model.add(Dense(n_units[i], activation='relu'),
#                      kernel_regularizer=tf.keras.regularizers.L1(0.01),
#                      activity_regularizer=tf.keras.regularizers.L2(0.01)
                     )
    model.add(Dense(1))
    return model
```

```python
def build_compile_model(n_hidden_layers, n_units, lr):
    model = build_NN(n_hidden_layers, n_units)
    model.compile(
        optimizer=tfo.Adam(learning_rate=lr),
        loss='mse',
        metrics=['RootMeanSquaredError']
    )
    model.summary()
    return model
```

```python
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import TimeSeriesSplit, GroupKFold
from sklearn.metrics import mean_squared_error
from datetime import datetime
import optuna
```

```python
train_eng = preprocess_data(train_data)
test_eng = preprocess_data(test_data)
X = train_eng.drop(['emission','ID_LAT_LON_YEAR_WEEK', 'latitude', 'longitude'], axis=1).fillna(0)
y = train_eng['emission']
features = X.columns.to_list()
target = 'emission'
```

```python
def predict_plot(model, loss_correction):
    scaler = StandardScaler()
    scaler.fit(X)
    example_df = pd.concat([train_eng, test_eng])
    example_df[target] = model.predict(scaler.transform(example_df[features]))
    example_df = example_df.groupby(example_df['year']).mean()
    plt.figure(figsize=(20, 6))
    plt.plot(example_df.index, example_df.emission*loss_correction, label='prediction')
    ori_df = train_data.copy()
    ori_df = ori_df.groupby('year').mean()
    plt.plot(ori_df.index, ori_df.emission, label='true', alpha=0.5, color='red')
    plt.legend()
    plt.title('All Predictions and true carbon emission for 4 years')
    plt.show()
```

```python
def eval_model(y_val, y_hat):
    plt.figure(figsize=(6, 6))
    plt.scatter(y_val, y_hat, s=1, color='r')
    plt.plot([plt.xlim()[0], plt.xlim()[1]], [plt.xlim()[0], plt.xlim()[1]], '--', color='k')
    plt.gca().set_aspect('equal')
    plt.xlabel('y_true')
    plt.ylabel('y_pred')
    plt.title('OOF Predictions')
    plt.show()
    
    residuals = y_hat - y_val
    
    plt.figure(figsize=(18, 4))
    plt.scatter(np.arange(len(residuals)), residuals, s=1)
    plt.title('All residuals by row number')
    plt.ylabel('residual')
    plt.show()
    
    plt.figure(figsize=(18, 4))
    plt.hist(residuals, bins=200)
    plt.title('Histogram of all residuals')
    plt.show()
    print(f"Standard deviation of log residuals: {residuals.std():.3f}")
    plt.show()
```

```python
def train_model(model, params, evaluate=False):
    scores = []
    start_time = datetime.now()
    for fold,i in enumerate(range(2021,2018,-1)):
        val = train_eng.loc[train_eng.year==i]
        train = train_eng.loc[train_eng.year!=i]
        print('Training on these years', set(train.year))
        print('Validating on these years', set(val.year)) 
        X_train = train[features]
        y_train = train[target]
        X_val = val[features]
        y_val = val[target]
        scaler = StandardScaler()
        X_tr_scaled = scaler.fit_transform(X_train)
        X_val_scaled = scaler.transform(X_val)

        if model=='cbr':
            model = CatBoostRegressor(**params, random_state=5, verbose=0,
                                      early_stopping_rounds=20, n_estimators=500)
            model.fit(X_tr_scaled, y_train, 
                      eval_set=(X_val_scaled, y_val)
                     )
        elif model=='xgb':
            model = XGBRegressor(**params, random_state=5, 
                                 objective='reg:squarederror', 
                                 early_stopping_rounds=20, n_estimators=500,
                                 n_jobs=-1, eval_metric='rmse')
            model.fit(X_tr_scaled, y_train, eval_set=[(X_val_scaled, y_val)])
        elif model=='lgbm':
            model = LGBMRegressor(**params, random_state=5, verbose=0, 
                                  num_boost_round=100,
                                  n_jobs=-1,
                                  early_stopping_round=100, n_estimators=1000,
                                  objective='regression_l1',
                                  )
            model.fit(X_tr_scaled, y_train, eval_set=[(X_val_scaled, y_val)])


        y_hat = model.predict(X_val_scaled)
        rmse = np.sqrt(mean_squared_error(np.array(y_val),np.array(y_hat)))
        print(f"Fold {fold} | {str(datetime.now() - start_time)[-12:-7]}"
              f" | RMSE: {rmse:.5f}")
        print()
    #      
        scores.append(rmse)
        if (i==2021)&(evaluate):
            eval_model(y_val, y_hat)

    return np.mean(scores), model
```

```python
def run_model(model, n_trials=20):
    def objective(trial):
        # Define the hyperparameter search space
        if model=='cbr':
            params = {
                'learning_rate': trial.suggest_float('learning_rate', 0.001, 0.2, log=True),
                'depth': trial.suggest_int('depth', 2, 10),
                'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1e-3, 1e2, log=True),
                'border_count': trial.suggest_int('border_count', 32, 255),
                'bagging_temperature': trial.suggest_float('bagging_temperature', 0.5, 5, log=True)
            }
        if model=='xgb':
            params = {
                'learning_rate': trial.suggest_float('learning_rate', 0.001, 0.1, log=True),
                'max_depth':trial.suggest_int('max_depth', 2, 10),
                'gamma': trial.suggest_float('gamma', 1e-4, 0.5, log=True),
                'subsample': trial.suggest_float('subsample', 0.6, 1.0),
                'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
                'reg_alpha': trial.suggest_float('reg_alpha', 1e-9, 10, log=True),
                'reg_lambda': trial.suggest_float('reg_lambda', 1e-9, 10, log=True),
            }
        if model=='lgbm':
            params = {
            'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.2, log=True),
            'num_leaves': trial.suggest_int('num_leaves', 10, 100),
            'max_depth': trial.suggest_int('max_depth', 3, 10),
            'min_child_samples': trial.suggest_int('min_child_samples', 1, 100),
            'feature_fraction': trial.suggest_float('feature_fraction', 0.6, 1.0),
            'bagging_fraction': trial.suggest_float('bagging_fraction', 0.6, 1.0),
            'bagging_freq': trial.suggest_int('bagging_freq', 1, 10),
            'lambda': trial.suggest_float('lambda', 1e-5, 10, log=True),
            'alpha': trial.suggest_float('alpha', 1e-5, 5, log=True),
        }
        rmse, _ = train_model(model, params)
        return rmse
    study = optuna.create_study(direction='minimize')
    study.optimize(objective, n_trials=n_trials)
    best_params = study.best_params
    best_score = study.best_value
    print('The best parameters are:', best_params)
    print('The best RMSE is:', best_score)
    return best_params
```

```python
best_params = run_model('xgb', 10)
```
