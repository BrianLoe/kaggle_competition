## Creating functions helper

```python
from tqdm import tqdm
def build_model(model, n, X, y):
    accuracy = []
    roc = []
    y_hat_l = []
    y_val_l = []
    pos_probs_l = []
    skf = StratifiedKFold(n_splits=n, shuffle=True, random_state=2) 
    for train_index, val_index in tqdm(skf.split(X, y)):
        X_train, X_val = X.loc[train_index,:], X.loc[val_index,:]
        y_train, y_val = y[train_index], y[val_index]
        model.fit(X_train, y_train)
        y_val_l.append(y_val)
        pos_probs = model.predict_proba(X_val)[:, 1]
        y_hat = (pos_probs > 0.4).astype('float')
        y_hat_l.append(y_hat)
        accuracy.append(accuracy_score(y_hat, y_val))
        pos_probs_l.append(pos_probs)
        roc_score = roc_auc_score(y_val, pos_probs)
        roc.append(roc_score)
    return accuracy, roc, y_val_l, y_hat_l, pos_probs_l
```

```python
def plot_cm(y_val, y_hat):
    matrix = confusion_matrix(y_val, y_hat)
    ax = sns.heatmap(matrix, annot=True, fmt='d', cmap='crest')
    ax.set(xticklabels=['Did not fail', 'Fail'], yticklabels=['Did not fail', 'Fail'],
           ylabel='Predicted values', xlabel='Actual values')
    plt.show()
```

```python
def plot_roc_curve(y_val, positive_probs):
    fpr, tpr, _ = roc_curve(y_val.astype('float'), positive_probs)
    plt.plot([0, 1], [0, 1], linestyle='--', label='No Algorithm')
    plt.plot(fpr, tpr, marker='.', label='Logistic')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.show()
```

```python
def build_evaluate_model(model, n, X, y):
    accuracy, roc, y_val_l, y_hat_l, pos_probs_l = build_model(model, n, X, y)
    best_index = np.argmax(accuracy)
    print('The highest accuracy is ',accuracy[best_index])
    print('The highest ROC score is', roc[best_index])
    y_val = y_val_l[best_index]
    y_hat = y_hat_l[best_index]
    pos_probs = pos_probs_l[best_index]
    print('Classification scores:')
    print(classification_report(y_val, y_hat))
    # confusion matrix
    plot_cm(y_val, y_hat)
    # roc auc
    plot_roc_curve(y_val, pos_probs)
```

# Neural Networks with Pytorch

```python
import torch
import torch.nn as nn
```

```python
class BinaryRNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(BinaryRNN, self).__init__()
        self.hidden_size = hidden_size
        self.rnn = nn.RNN(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        h0 = torch.zeros(1, x.size(0), self.hidden_size).to(x.device)
        out, _ = self.rnn(x, h0)
        out = self.fc(out[:, -1, :])
        return out
```

```python
# split to train and validation set
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=.2, random_state=2)
```

```python
# initialise parameters
input_size = len(cols)
hidden_size = 128
output_size = 1
num_epochs = 100
```

```python
# build model
model = BinaryRNN(input_size, hidden_size, output_size)

criterion = nn.BCEWithLogitsLoss() # set loss function binary cross-entropy
optimizer = torch.optim.Adam(model.parameters(), lr=0.06)
inputs = torch.tensor(X_train.astype('float').values, dtype=torch.float32).unsqueeze(1) # set inputs
label = torch.tensor(y_train.astype('float').values, dtype=torch.float32) # set target outputs
# train model
for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(inputs)
    loss = criterion(outputs.squeeze(), label)
    loss.backward()
    optimizer.step()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
```

```python
# evaluate model
model.eval()
with torch.no_grad():
    y_pred = model(torch.tensor(X_val.astype('float').values, dtype=torch.float32).unsqueeze(1))
    y_pred_prob = torch.sigmoid(y_pred)
    y_pred_prob_np = y_pred_prob.cpu().numpy().squeeze()
    y_val_np = np.array(y_val.astype('float'))
    roc_auc = roc_auc_score(y_val_np, y_pred_prob_np)
    predicted_labels = (y_pred.sigmoid() > 0.5).float()
```

```python
y_val_np = np.array(y_val.astype('float'))
y_hat = np.array(predicted_labels.squeeze())
accuracy = accuracy_score(y_val_np, y_hat)
print('Accuracy score is', accuracy)
```

```python
print('ROC score is', roc_auc)
```

```python
print(classification_report(np.array(y_val.astype('float')), np.array(predicted_labels.squeeze())))
```

```python
plot_cm(y_val_np, y_hat)
```

```python
plot_roc_curve(y_val_np, y_pred_prob_np)
```
